INNER JOIN AND LEFT OUTER JOIN IN SPARK

```python
visits = sc.parallelize([("h", "1.2.3.4"), ("a", "3.4.5.6"), ("h","1.3.3.1")] )
pageNames = sc.parallelize([("h", "Home"), ("a", "About"), ("o", "Other")])
cg = visits.cogroup(pageNames).map(lambda x :(x[0], ( list(x[1][0]), list(x[1][1]))))
```

```python
cg.collect()
```

```python
def J(x):
    j=[]
    k=x[0]
    if x[1][0]!=[] and x[1][1]!=[]:
        for l in x[1][0]:
            for r in x[1][1]:
                j.append((k,(l,r)))
    return j
```

```python
innerjoin = cg.flatMap(lambda x: J(x))
```

```python
innerjoin.collect()
```

```python
def roJ(x):
    j=[]
    k=x[0]
    if x[1][0]!=[] and x[1][1]!=[]:
        for l in x[1][0]:
            for r in x[1][1]:
                j.append((k,(l,r)))
    elif x[1][1]!=[] :
        for r in x[1][1]:
            j.append((k, (None, r)))
    return j
```

```python
rightouterjoin = cg.flatMap(lambda x: roJ(x))
```

```python
rightouterjoin.collect()
```
